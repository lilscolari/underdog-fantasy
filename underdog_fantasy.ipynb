```python
import nba_api
import pandas as pd
from nba_api.stats.endpoints import commonplayerinfo
import numpy as np
```

```python
from nba_api.stats.endpoints import leaguegamefinder

gamefinder = leaguegamefinder.LeagueGameFinder()
games = gamefinder.get_data_frames()[0]

games_2024 = games[games.SEASON_ID.str[-4:] == '2024']
games_2024
```

```python
team_codes = ['SAS', 'NOP', 'ATL', 'BOS', 'CHA', 'UTA', 'CHI', 'CLE', 'DAL', 'DEN', 'DET', 'GSW', 'HOU', 'IND', 'LAC', 'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NYK', 'BKN', 'OKC', 'ORL', 'PHI', 'PHO', 'POR', 'SAC', 'TOR', 'WAS']
len(team_codes)
```

```python
games = games_2024[games_2024['TEAM_ABBREVIATION'].apply(lambda x: x in team_codes)]
filtered_games = games[(games['GAME_DATE'] >= '2024-10-22') & (games['GAME_DATE'] != '2024-12-17')]
filtered_games = filtered_games.drop_duplicates(subset='GAME_ID').reset_index()
filtered_games
```

```python
def calculate_fantasy_points(d):
    assists = int(d['assists'])
    blocks = int(d['blocks'])
    points = int(d['points'])
    steals = int(d['steals'])
    turnovers = int(d['turnovers'])
    rebounds = int(d['reboundsTotal'])
    return points + (assists * 1.5) + (rebounds * 1.2) + ((blocks + steals) * 3) - turnovers
```

```python
def input_player(player):
    stats = player['statistics']
    name = player['name']
    fp = calculate_fantasy_points(stats)
    if name in all_players.keys():
        all_players[name]['total'] += fp
        if fp > all_players[name]['max']:
            all_players[name]['max'] = fp
        if fp < all_players[name]['min']:
            if fp == 0.0:
                break
            all_players[name]['min'] = fp
    else:
        all_players[name] = {'total': fp, 'max': fp, 'min': fp}
```

```python
all_players = {}
```

```python
from nba_api.live.nba.endpoints import boxscore

for i, game_id in enumerate(filtered_games['GAME_ID']):
    print(f'Added game {i}')
    boxscore_data = boxscore.BoxScore(game_id)
    
    game_data = boxscore_data.game.get_dict()    
    
    for player in game_data['homeTeam']['players']:
        input_player(player)
    for player in game_data['awayTeam']['players']:
        input_player(player)
```

```python
fantasy_df = pd.DataFrame.from_dict(all_players, orient='index')
fantasy_df = fantasy_df.reset_index()
fantasy_df = fantasy_df.rename(columns={'index': 'Player_Name'})
```

```python
pd.set_option('display.max_rows', None)

# Now print the DataFrame
print(fantasy_df.sort_values(by='total', ascending=False).reset_index())

# Reset to default after use (optional)
pd.reset_option('display.max_rows')
```
